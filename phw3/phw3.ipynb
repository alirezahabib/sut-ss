```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1

```python
def x1(t):
    return np.vectorize(lambda x : 0 if (x > 3 or x < 0) else 1)(t)
```

## a
Compute the discrete time Fourier transform and plot its magnitude and
phase.

```python
N = 1000
t = np.arange(N)
x_data = x1(t)
f = np.arange(N)*100/N

X = np.fft.fft(x_data)
X_mag = np.abs(X)
X_angle = np.angle(X)

plt.plot(f, X_mag)
plt.xlabel('f')
plt.ylabel('Magnitude')
plt.show()

plt.plot(f, X_angle)
plt.xlabel('f')
plt.ylabel('Phase')
plt.show()
```

## b
Compute the 4-point DFT of x(n).

```python
N = 4
t = np.arange(N)
x_data = x1(t)
f = np.arange(N)*100/N

X = np.fft.fft(x_data)
X_mag = np.abs(X)
X_angle = np.angle(X)

plt.plot(f, X_mag)
plt.xlabel('f')
plt.ylabel('Magnitude')
plt.show()

plt.plot(f, X_angle)
plt.xlabel('f')
plt.ylabel('Phase')
plt.show()
```

## c
How can we obtain other samples of the DTFT?

We can use other values for N (other than 1000, which is used up there), or we can shift our sampling interval to start from something different than 0.
But we should include our 4 points in the sampling; otherwise, nothing would be left rather than a zero function.

# 2

```python
def x2(t):
    return np.vectorize(lambda x : (2*np.exp(-0.9*np.absolute(x))) if (-5 <= x and x <= 5) else 0.0)(t)
```

## a
Plot the DTFT of the above sequence using DFT as a computation
tool. Choose the length N of the DFT so that this plot appears to be a smooth
graph

```python
N = 100
t = np.linspace(-N+1, N, 2*N)  
x_data = x2(t)
f = np.arange(2*N)*100/N/2

X = np.fft.fft(x_data)
X_mag = np.absolute(X)
X_angle = np.angle(X)

plt.plot(f, X_mag)
plt.xlabel('f')
plt.ylabel('Magnitude')
plt.show()

plt.plot(f, X_angle)
plt.xlabel('f')
plt.ylabel('Phase')
plt.show()
```

```python
def plot_fft(N, func):
    t = np.linspace(-N+1, N, 2*N)  
    x_data = func(t)
    f = np.arange(2*N)*100/N/2

    plt.plot(f, np.absolute(np.fft.fft(x_data)))
    plt.xlabel('f')
    plt.ylabel('Magnitude')
    plt.show()
```

```python
def x3_1(n):
    return np.power(0.6, np.absolute(n))*(np.heaviside(n + 10, 0) - np.heaviside(n - 11, 0))
```

```python
def x3_2(n):
    return n*np.power(0.9, n)*(np.heaviside(n, 0) - np.heaviside(n - 21, 0))
```

```python
def x3_3(n):
    return (np.cos(0.5*np.pi*n) + 1j*np.sin(0.5*np.pi*n))*(np.heaviside(n, 0) - np.heaviside(n - 51, 0))
```

```python
def x3_4(n):
    return np.vectorize(lambda x : max(0, 4 - np.absolute(x)))(n)
```

```python
def x3_5(n):
    return np.sign(n) * x3_4(n)
```
